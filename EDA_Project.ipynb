**Histogram**

```python
plt.figure(figsize=(8,5))
survey_df['AI_Replacement_Risk'].hist()
plt.title("Distribution of Automation Risk")
plt.show()
```

**Count Plot**

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Industry',data=survey_df)
plt.xticks(rotation=45)
plt.show()
```

**Bar Plot**

```python
plt.figure(figsize=(10,5))
sns.barplot(
    x='Industry',
    y='AI_Tool_Usage',
    data = survey_df
)
plt.xticks(rotation=45)
plt.show()
```

**Scatter Plot**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='AI_Tool_Usage',
    y='Job_Growth_2030',
    data=survey_df
)
plt.show()
```

**Heatmap**

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    survey_df.select_dtypes(include=np.number).corr(),
    annot=True
)
plt.show()
```

Q1. Which industry has the highest Ai Adaptation ?

```python
usage_mapping = {'Low': 0, 'Moderate': 1, 'High': 2}
survey_df['AI_Tool_Usage_Numeric'] = survey_df['AI_Tool_Usage'].map(usage_mapping)
survey_df.groupby('Industry')['AI_Tool_Usage_Numeric'].mean().sort_values(ascending=False)
```

Q2. Which industry faces the highest automation risk ?

```python
survey_df.groupby('Industry')['AI_Replacement_Risk'].mean().sort_values(ascending=False)
```

Q3. Does higher AI adoption lead to job growth ?

```python
survey_df[['AI_Tool_Usage_Numeric',
           'Job_Growth_2030']].corr()
```

Q4. Which industry has highest salary growth ?

```python
survey_df.groupby('Industry')['Average_Salary_USD'].mean().sort_values(ascending=False)
```

Q5. which jobs are safest from automation ?

```python
survey_df.sort_values(
    by='AI_Replacement_Risk'
).head(10)
```
