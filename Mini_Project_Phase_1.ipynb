```python
# Project:-2
# COVID-19 Data Analysis
# Objective:
# Analyze COVID cases across countries/states.
# Features:
# •	Daily case trends
# •	Recovery rate
# •	Death rate
# •	Country-wise comparison
# Concepts:
# •	Time Series Analysis
# •	Data Filtering
# •	Statistical Analysis


import pandas as pd
import numpy as np

df = pd.read_excel("covid_19_clean_complete.csv.xlsx")

# Display Dataset
print("First 5 Rows:")
print(df.head())

print("\nLast 5 Rows:")
print(df.tail())

# Data Cleaning
df["Province/State"] = df["Province/State"].fillna("Unknown")
df["Date"] = pd.to_datetime(df["Date"])


# Feature 1 : Daily Case Trends

daily_cases = df.groupby("Date")["Confirmed"].sum()

print("\nDaily Case Trends:")
print(daily_cases)

# Feature 2 : Recovery Rate
# Formula : Recovery Rate = Recovered/Confirmed*100

latest = df.groupby("Country/Region").last()

latest["Recovery Rate"] = (
    latest["Recovered"] /
    latest["Confirmed"]
) * 100

print("\nTop Countries by Recovery Rate:")
print(
    latest[["Recovery Rate"]]
    .sort_values("Recovery Rate", ascending=False)
    .head(10)
)

# Feature 3 : Death Rate
# Formula: Death Rate = Deaths/Confirmed * 100

latest["Death Rate"] = (
    latest["Deaths"] /
    latest["Confirmed"]
) * 100


print("\nTop Countries by Death Rate:")
print(
    latest[["Death Rate"]]
    .sort_values("Death Rate", ascending=False)
    .head(10)
)

# Feature 4 : Country Wise comparision
top_countries = (
    latest.sort_values(
        "Confirmed",
        ascending = False
    )
    .head(10)
)

print("\nTop 10 Countries by Confirmed Cases:")
print(top_countries["Confirmed"])

# Data Filtering :
# Example:
# India :

india = df[df["Country/Region"] == "India"]

print("\nIndia Data:")
print(india.head())

# US :

us = df[df["Country/Region"] == "US"]

print("\nUS Data:")
print(us.head())


# Data For 2020

data_2020 = df[df["Date"].dt.year == 2020]

print("\nData for 2020:")
print(data_2020.head())

# Statistical Analysis

print("\nMean Confirmed Cases:")
print(df["Confirmed"].mean())

print("\nMedian Confirmed Cases:")
print(df["Confirmed"].median())

print("\nStandard Deviation:")
print(df["Confirmed"].std())

# Correlation b/w Confirmed, Deaths, Recovered

corr = df[
    ["Confirmed", "Deaths", "Recovered"]
].corr()

print("\nCorrelation Matrix:")
print(corr)

# Most Affected Countries

print("\nMost Affected Countries:")
print(
    latest.sort_values(
        "Confirmed",
        ascending=False
    ).head(10)
)

# Highest Recovery Rate

print("\nHighest Recovery Rate:")
print(
    latest.sort_values(
        "Recovery Rate",
        ascending = False
    ).head(10)
)

# Highest Death Rate

print("\nHighest Death Rate:")
print(
    latest.sort_values(
        "Death Rate",
        ascending = False
    ).head(10)
)
```
